```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for column in df.columns.values:
  if df[column].dtypes=="object":
    data = df[column].append(df[column])
    le.fit(data.values)
    df[column] = le.transform(df[column])
df
```
